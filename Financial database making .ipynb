```python
#Importing The Packages

import numpy as np
import random
import pandas as pd
import string
from faker import Faker
```

```python
# Creting the Tabel Name :- Customer Details

n=300

#Customer_id column generation
def customer_id():
    digits = string.digits
    cust_id = [f'10' + ''.join(np.random.choice(list(digits), 5)) for i in range(n)]
    return cust_id

customer = customer_id()


#Customer Name generation
fake = Faker()
Customer_name = [fake.name() for i in range(n)]
name = Customer_name


# Date of Birth generation
year = np.random.randint(2000, 2005, n)
month = np.random.randint(1, 13, n)
day = np.random.randint(1, 30, n)

date_birth = [f'{year[i]}-{str(month[i]).zfill(2)}-{str(day[i]).zfill(2)}' for i in range(n)]
dob = date_birth


# Email ids generation
def generate_random_email(name):
    domain = ['gmail.com', 'ac.co.uk', 'yahoomail.com', 'rediffmail.com','msn.com','techno.co.in','nebulafox.co']
    domains = np.random.choice(domain)
    return f"{name.replace(' ', '.').lower()}@{domains}"

random_emails = [generate_random_email(name) for name in name]


#Address generation
def fake_address():
    fake = Faker('en-US')
    fk_address = fake.address()
    return fk_address

addresses = [fake_address() for i in range(n)]


#city generation
def City():
    fake = Faker('en-US')
    fk_city = fake.city()
    return fk_city
city = [City()for i in range(n)]






```

```python
#Creating the Dataframe
df = pd.DataFrame({
    'Customer_ID': customer,
    'Customer Name': name,
    'Email-ID': random_emails,
    'Date of Birth':dob,
    'Address': addresses,
    'City':city
    
})

#Converting the Dataframe into .CSV file
financial_database = df[['Customer_ID','Customer Name','Email-ID','Date of Birth','Address','City']]
financial_database.to_csv('CustomerDetails.csv')
```

```python
# Creting the Tabel Name :- Account Details

#1 #Customer_ids
customer

# #2 Account Number or ID
def Account_number(length=7):
    number = string.digits
    account_num = ''.join(np.random.choice(list(number), length))
    return account_num
Account_ID = [Account_number() for i in range(n)]


#  #3Account Type
def Account_type():
    Account_categories = ['Saving Account','Current Account','Business Account','Student Account']
    cate = np.random.choice(Account_categories)
    return cate
account_type = [Account_type() for i in range(n)]


# 4 Account OpenDate
year = np.random.randint(2000, 2010, n)
month = np.random.randint(1, 13, n)
day = np.random.randint(1, 30, n)

account_opendate = [f'{year[i]}-{str(month[i]).zfill(2)}-{str(day[i]).zfill(2)}' for i in range(n)]


# 5 Customer_Balance
def Customer_balance():
    min_balance = 50000
    max_balance = 500000
    balance = np.random.randint(min_balance,max_balance)
    return balance
Balance = [Customer_balance()for i in range(n)]



```

```python
#creating the dataframe
df1 = pd.DataFrame({
    'Customer ID': customer,
    'Account Number': Account_ID,
    'Account Type':account_type,
    'Account Open Date':account_opendate,
    'Account Balance':Balance 
    
})

print(df1.head())
# converting the dataframe into .csv
Account_info = df1[['Customer ID','Account Number','Account Type', 'Account Open Date','Account Balance']]
Account_info.to_csv('AccountDetails.csv')
```

```python
# Creting the Tabel Name :- Transaction Details

#1 #Customer_ids
customer


# # 2 A/C ID:
Account_ID


# #3 Transaction ID
def Transaction_ID(length=9):
    characters = string.ascii_lowercase + string.digits
    tran_id = ''.join(random.choice(characters) + str(random.randint(0, 9)) for _ in range(length//2))
    return tran_id
ids = [Transaction_ID()for i in range(n)]
    
    
# # 4 Transaction Date
year = np.random.randint(2000, 2018, n)
month = np.random.randint(1, 13, n)
day = np.random.randint(1, 30, n)

Transaction_Date = [f'{year[i]}-{str(month[i]).zfill(2)}-{str(day[i]).zfill(2)}' for i in range(n)]



# # 5 Transaction Type
transaction_type = ['Deposit','Withdrawl','Transfer','Payment','Purchase']
tran_type = [np.random.choice(transaction_type) for i in range(n)]





# # 6 Transaction Amount
def Transaction_amount():
    min_balance = 50000
    max_balance = 500000
    tran_amt = np.random.randint(min_balance, max_balance)
    return tran_amt
Amount=[Transaction_amount()for i in range(n)]


```

```python
#Creating The Dataframe
df2 = pd.DataFrame({
    'Customer_ID': customer,
    'Account_Number': Account_ID,
    'Transaction ID': ids,
    'Transaction Date':Transaction_Date,
    'Transaction Type':tran_type,
    'Transaction Amount':Amount
    
    
})

# Converting the Dataframe .CSV
financial_database = df2[['Customer_ID','Account_Number','Transaction ID','Transaction Date','Transaction Type','Transaction Amount']]
financial_database.to_csv('TransactionHistory.csv')
```

```python
# Creting the Tabel Name :- Loan Details

# 1.Customer ID
customer

#2.Customer Name
name

#3 Account ID
Account_ID



#3.Loan Type
def generate_random_loan():
    loan_type = random.choice(['Personal Loan', 'Home Loan', 'Auto Loan', 'Student Loan', 'Business Loan'])
    return loan_type
Loan = [generate_random_loan() for i in range(n)]


# 4 Loan Amount
def Loan_amount():
    amount = round(random.uniform(100000, 2500000), 2)
    return amount
loan_amt = [Loan_amount()for i in range(n)]



# 5Interest rate
def generate_random_loan():
    loan_type = random.choice(['Personal Loan', 'Home Loan', 'Auto Loan', 'Student Loan', 'Business Loan'])
    if loan_type == 'Personal Loan':
        interest_rate = round(random.uniform(10.38, 13.35), 2)  
    elif loan_type == 'Home Loan':
        interest_rate = round(random.uniform(6.5, 8.5), 2) 
    elif loan_type == 'Auto Loan':
        interest_rate = round(random.uniform(7.5, 11.5),2)
    elif loan_type == 'Student Loan':
        interest_rate = round(random.uniform(6.5, 11.45), 2) 
    elif loan_type == 'Business Loan':
        interest_rate = round(random.uniform(9.25, 18.86), 2)  
    else:
        interest_rate = 10.00  # Default interest rate

    return interest_rate

loan_data = [generate_random_loan() for i in range(n)]

# 6 Duration month
def Duration_month():
    duration = np.random.randint(18, 90)
    return duration
durations = [Duration_month() for i in range(n)]

```

```python
# Creating the DataFrame
df3 = pd.DataFrame({
    'Customer_ID': customer,
    'Account Number': Account_ID,
    'Loan Type':Loan,
    'Loan Amount':loan_amt,
    'Interest Rate':loan_data,
    'Duration':durations
    
    
})

# Converting the Dataframe to .CSV file
financial_database = df3[['Customer_ID','Account Number','Loan Type','Loan Amount','Interest Rate','Duration']]
financial_database.to_csv('LoanDetails.csv')
```
